```python
import sqlite3
import pandas as pd

con = sqlite3.connect("dataset_final.db")
def q(sql): return pd.read_sql_query(sql, con)
```

```python
# De fösta 10 bilar
q("""
SELECT url, regnr, model_year, price_sek, odometer_km, fuel, body_type, price_per_1000km
FROM fact_cars LIMIT 10;
""")
```

```python
#Totaler och fördelning
display(q("SELECT COUNT(*) AS rows FROM fact_cars;"))
display(q("""
SELECT fuel, COUNT(*) AS n, ROUND(AVG(price_sek),0) AS avg_price
FROM fact_cars GROUP BY fuel ORDER BY n DESC;
"""))
```

```python
#Modellår och medelpris
q("""
SELECT model_year, COUNT(*) AS n, ROUND(AVG(price_sek),0) AS avg_price
FROM fact_cars
WHERE model_year IS NOT NULL AND model_year >= strftime('%Y','now') - 15
GROUP BY model_year
ORDER BY model_year;
""")
```

```python
#Spara utdrag till csv
q("""
SELECT regnr, model_year, fuel, body_type, price_sek, odometer_km, price_per_1000km, url
FROM fact_cars
""").to_csv("export_fact_cars.csv", index=False, encoding="utf-8")
```
